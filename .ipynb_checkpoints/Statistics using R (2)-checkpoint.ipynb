# 常见分布

## 正态分布 (Normal distribution)
<br/>
<font size = 3>
    
**正态分布** 又称高斯分布 (Gaussian distribution)，是一个在数学、物理及工程等领域都非常重要的概率分布，在统计学的许多方面有着重大的影响力。

随机变量 $X$服从正态分布，记为 $X \sim N(\mu, \sigma^2)$。概率密度函数为正态分布的期望值决定了其位置，其方差决定了分布的幅度。偏度系数和峰度系数都是 0，特征函数为$\Phi(t) = exp(it\mu-\sigma^2t^2/2)$ 

概率密度函数：

$f(x;\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma}}exp(-\frac{(x-\mu)^2}{2\sigma^2}), \text{其中}\mu \in R,\sigma>0$
    
累计分布函数：
$F(x;\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma}}\int _{-\infty} ^{x}exp(-\frac{(t-\mu)^2}{2\sigma^2})$

数学期望为$E(X) = \mu$、方差为 $V(X) = \sigma^2$

    
当$\mu=0$,$\sigma=1$时，$X ∼ N(0, 1)$称为**标准正太分布**,其概率密度函数为：

$f(x) =\frac{1}{\sqrt{2\pi}}exp(-\frac{x^2}{2})$

正太分布$X ∼ N(\mu, \sigma^2)$与标准正太分布$X ∼ N(0, 1)$有如下关系：

若$ Y=\frac{X-\mu}{\sigma}$,
则 $Y\sim N(0,1)$,该变化称为**X的标准化**

```r
#加载必要的库
library(ggplot2)
```

```r
#正太分布概率密度函数
fun1 <- function(x){
y <- dnorm(x,mean = 0,sd = 1)
return(y)
}
fun2 <- function(x){
y <- dnorm(x,mean = 0,sd = 0.5)
return(y)
}
fun3 <- function(x){
y <- dnorm(x,mean = 0,sd = 2)
return(y)
}
ggplot(NULL, aes(x=x, colour = g)) +
stat_function(data = data.frame(x = -5:5, g = factor(1)), fun = fun1) +
stat_function(data = data.frame(x = -5:5, g = factor(2)), fun = fun2) +
stat_function(data = data.frame(x = -5:5, g = factor(3)), fun = fun3) +
scale_colour_manual(values = c("red", "green","blue"),
labels = c("red","green","blue"))+
ylab(label = "density")+
labs(title="The Normal Density Distribution")
```

```r
#正太累积分布
fun1 <- function(x){
y <- pnorm(x,mean = 0,sd = 1)
return(y)
}
fun2 <- function(x){
y <- pnorm(x,mean = 0,sd = 0.5)
return(y)
}
fun3 <- function(x){
y <- pnorm(x,mean = 0,sd = 2)
return(y)
}
ggplot(NULL, aes(x=x, colour = g)) +
stat_function(data = data.frame(x = -5:5, g = factor(1)), fun = fun1) +
stat_function(data = data.frame(x = -5:5, g = factor(2)), fun = fun2) +
stat_function(data = data.frame(x = -5:5, g = factor(3)), fun = fun3) +
scale_colour_manual(values = c("red", "green","blue"),
labels = c("red", "green","blue"))+
ylab(label = "density")+
labs(title="The Normal Cumulative Distribution")
```

## 指数分布 (Exponential distribution)
<br/>
<font size =3>
    
**指数分布 (Exponential distribution)** 用来表示独立随机事件发生的时间间隔。指数分布是伽玛分布和 weibull 分布的特殊情况。记为$X \sim Expon(\lambda) ,\lambda$为参数。其概率密度函数为：

$f(x;\lambda) = \left\{
\begin{aligned}
&\lambda e^{-\lambda x}&\text{当}x\gt0\\ \\
&0 &\text{当}x\leq0
\end{aligned}
\right.$ 其中$\lambda \gt 0$
    
累计分布函数：

$F(x;\lambda) = \left\{
\begin{aligned}
&1-e^{\lambda x}&\text{当}x\gt0\\ \\
&0 &\text{当}x\leq0
\end{aligned}
\right.$ 其中$\lambda \gt 0$
  
数学期望为$E(X) = 1/\lambda$、方差为 $V(X) = \frac{1}{\lambda^2}$

```r
#指数分布概率密度函数
fun1 <- function(x){
y <- dexp(x,0.5)
return(y)
}
fun2 <- function(x){
y <- dexp(x,1)
return(y)
}
fun3 <- function(x){
y <- dexp(x,2)
return(y)
}
ggplot(NULL, aes(x=x, colour = g)) +
stat_function(data = data.frame(x=seq(0,3,length.out=100),
g = factor(1)), fun = fun1) +
stat_function(data = data.frame(x=seq(0,3,length.out=100),
g = factor(2)), fun = fun2) +
stat_function(data = data.frame(x=seq(0,3,length.out=100),
g = factor(3)), fun = fun3) +
scale_colour_manual(values = c("red", "green","blue"),labels = c("red", "green","blue"))+
ylab(label = "density")+
labs(title="The Exponential Density Distribution")
```

```r
#指数累积分布
fun1 <- function(x){
y <- pexp(x,0.5)
return(y)
}
fun2 <- function(x){
y <- pexp(x,1)
return(y)
}
fun3 <- function(x){
y <- pexp(x,2)
return(y)
}
ggplot(NULL, aes(x=x, colour = g)) +
stat_function(data = data.frame(x=seq(0,3,length.out=100),
g = factor(1)), fun = fun1) +
stat_function(data = data.frame(x=seq(0,3,length.out=100),
g = factor(2)), fun = fun2) +
stat_function(data = data.frame(x=seq(0,3,length.out=100),
g = factor(3)), fun = fun3) +
scale_colour_manual(values = c("red", "green","blue"),
labels = c("red", "green","blue"))+
ylab(label = "density")+
labs(title="The Exponential Cumulative Distribution Function")
```
