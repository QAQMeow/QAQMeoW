```python
# -*- coding: utf-8 -*-
#
import jieba
import pandas as pd
import path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import shutil
from wordcloud import WordCloud,STOPWORDS

#去重
file0 = 'F:/Jupyter/SnowPeach_Chan/Data/Bilibili/Bilibili_Barrage.csv'
csv = pd.read_csv(file0)#,engine='python')

data = csv.drop_duplicates(subset=None,keep='first',inplace=False).replace('\u2764','')

#data.to_csv(file2,encoding='utf-8-sig')
#词频统计

#de = pd.read_csv(file0,usecols=[1])
#dy = pd.read_csv(file0,usecols=[0],encoding='utf-8-sig')
#de.to_csv(file3,encoding = 'utf-8-sig')
#time = de.values
#test = dy.values
#lie = []
#for i in test:
 #   lie.append(i[0])
#print(lie)
#reult = pd.value_counts(lie)
#print(reult)

#制作词云

Pic0_path = 'F:/Jupyter/SnowPeach_Chan/Result/Pic0.png'
file = open(file0,encoding='utf-8-sig').read()
#print(file)
default_mode = jieba.cut_for_search(file)
text = " ".join(default_mode)

#mask = np.array(Image.open('F:/Jupyter/Untitled Folder/wordcloud/p.png'))
stopwords = set(STOPWORDS)
stopwords.add("said")
delet = ["了","啊","的","你","吧","我","是","这","好","就","这","不","都","吗","在","看","人",
         "也","有","还","呢","又","要","和","他","没","吃","被","不是","就是","什么","她","来",
         "能","真","呢","说","这个","真的","现在","这样","自己","还是","怎么","电影","那",
         "对","很","呀","太","哈哈","哈哈哈","没有","个","哈哈哈哈","一个","这么","得","想","让",
         "去","把","好看","哈","谁","上","啥","他们","真是","问题","豆腐","给","米","老婆","多","演技",
         "才","这种","那个","演","到","叫","过","李","组长","会","出来","时候","开始","这是","你们","么",
         "可以","卧槽","所以","不能","这女","有点","人家","做","里","啦","可","大","嘛","跟","着","咋","哪","写","像","啊啊啊"]
for i in delet:
    stopwords.add(i)
wc = WordCloud(
    font_path='F:/Jupyter/SnowPeach_Chan/文泉驿正黑.ttf',
    background_color="white",
    max_words=500,
    width = 1920,
    height = 1080,
    relative_scaling = 0.8,
    prefer_horizontal = 1,
   # mask=mask,
    stopwords=stopwords
)
#wc.fit_words(frequencies) 
wc.generate(text)
wc.to_file(Pic0_path)

#plt.imshow(wc,interpolation='bilinear')
#plt.axis("off")
#plt.figure()
#plt.imshow(cmap=plt.cm.gray,interpolation='bilinear')
#plt.axis("off")
#plt.show()

text2 = text.split(" ")
for i in range(len(text2)-1,0,-1):
    if text2[i] == '\n':
        del(text2[i])
l2 = ''
l3 = ''
l4 = ''
l5 = ''
l6 = ''
l7 = ''
l8 = ''
for i in text2:
    if len(i) == 2:
        l2+=(i+' ')
    if len(i) == 3:
        l3+=(i+' ')
    if len(i) == 4:
        l4+=(i+' ')
    if len(i) == 5:
        l5+=(i+' ')
    if len(i) == 6:
        l6+=(i+' ')
    if len(i) == 7:
        l7+=(i+' ')
    if len(i) == 8:
        l8+=(i+' ')

#wc = WordCloud(
#   font_path='F:/Jupyter/SnowPeach_Chan/文泉驿正黑.ttf',
#  background_color="white",
#    max_words=2000,
#   width = 1920,
#   height = 1080,
#  # mask=mask,
#   stopwords=stopwords
#)
p_dir = 'F:/Jupyter/SnowPeach_Chan/Result/'
d = ['L2.png','L3.png','L4.png','L5.png','L6.png','L7.png','L8.png']
s = [l2,l3,l4,l5,l6,l7,l8]
for i in range(0,len(s),1):
    wc.generate(s[i])
    wc.to_file(p_dir+d[i])
    
    
#
# for i in range(0,1000):
#     try:
#         dataframe = pd.DataFrame({'test':test,'time':time})
#
#     except:
#         continue

#d = path.dirname(__file__)
#path.dirname()



# dataframe = pd.DataFrame({'name'})
# words = open('qresult.csv',encoding='gbk').read()
# cloud = WordCloud(width=1000,height=860,margin=2,background_color='black')
#
# cloud.generate(words)
# cloud.to_file('qresult.jpg')
```

```python
lenMax = 0         #分词最长字数
for i in text2:
    if len(i)>=lenMax:
        lenMax = len(i)

print("最长字数 = ",lenMax)

lens = [0]*30
for i in range(0,30,1):
    for j in text2:
        if len(j) == i:
            lens[i]+=1
```

```python
X = [i for i in range(30)]
plt.figure(figsize=(64,40), dpi=25)
plt.title("Words Distribution",size = 90) 
plt.xlabel("Length of words",size = 70) 
plt.xticks(X,size = 50)
plt.yticks(size = 50)
plt.ylabel("number",size = 70) 
plt.ylim(0,40)
for x,y in zip(X[0:5],lens[0:5]):
     plt.text(x, 39, '%.0f'%y, ha='center', va= 'bottom',fontsize=33,alpha = 0.7)
for x,y in zip(X[5:30],lens[5:30]):
     plt.text(x, y%40, '%.0f'%y, ha='center', va= 'bottom',fontsize=33,alpha = 0.7)
plt.bar(X,lens,alpha=0.4,width =0.9,color = 'c')
plt.savefig("../Result/WordsDistribution.png")
plt.show()
```
